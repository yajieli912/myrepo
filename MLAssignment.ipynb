## 2. Writing an apply function that transforms a column

The function I wrote is to unify the input of Gender column

```python
def treatment_revise(treatment):
    if treatment == 'Yes':
        return 1
    else:
        return 0
df['treatment'] = df['treatment'].apply(treatment_revise)
df
```
